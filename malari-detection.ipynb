# Pre-processing 

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
help(ImageDataGenerator)
```

```python
# create object from ImageDataGenerator
train_datagen=ImageDataGenerator(
    rotation_range=20,# decide the angle to rotate the image
    width_shift_range=0.1,# ازاحه
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    channel_shift_range=0.0,
    fill_mode='nearest',# filling missing pixels with nearest pixels 
    horizontal_flip=True,#قلب الصوره افقيا
    #vertical_flip=True,#قلب الصوره عموديا
    rescale=1./255,# normalization( each image over 255)
    
)
```

```python
# image before augment
plt.imshow(uninfected_img_show)
```

```python
# using random_transform func, then plot it must previously output from imread
# show augmented image
plt.imshow(train_datagen.random_transform(uninfected_img_show))
```

```python
#image_shape=(imread(os.listdir(train_path +'/parasitized'+'/'+'C100P61ThinF_IMG_20150918_144104_cell_162.png'))).shape
image_shape = (130, 130, 3)
```

```python
#ig.fit(train_path)image_shape[0:2]# extract first two number from image_shape
train_gen= train_datagen.flow_from_directory(
    train_path,target_size=image_shape[0:2], color_mode='rgb', batch_size=16, class_mode='binary', shuffle=True)
test_gen= train_datagen.flow_from_directory(
    test_path,target_size=image_shape[0:2], color_mode='rgb', batch_size=16, class_mode='binary', shuffle=False)
```

```python
 train_gen.class_indices
```

```python
 test_gen.class_indices
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Flatten, Conv2D, Dense, MaxPooling2D,Dropout
```

```python
#Input_shape=(139, 160,3)
image_shape = (130, 130, 3)
```

```python
model=Sequential()
model.add(Conv2D(filters=32, kernel_size=(3,3), input_shape=image_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(filters=64, kernel_size=(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(filters=128, kernel_size=(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())

model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Training the Model
history =model.fit(train_gen, epochs=30,validation_data=test_gen)#, callbacks=[callback])
```

```python
losses=pd.DataFrame(model.history.history)
losses
```

```python
losses[['accuracy','val_accuracy']].plot()
```

```python
model.metrics_names
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred=model.predict(test_gen)
y_pred
```

```python
# Model evaluation
scores = model.evaluate(test_gen, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
